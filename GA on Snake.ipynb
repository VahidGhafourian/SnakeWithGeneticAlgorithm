```
class Snake :

    def __init__ (self, mapSize, name) :
        self.delay = 2
        self.snakeSize = 2 
        

        self.strRight = "R"
        self.strLeft = "L"
        self.strUP = "U"
        self.strDown = "D"
        self.strNothing = "N"
        self.dirr = self.strNothing
        self.name = name
        
# /**         -->>  set Snake first position   <<--  */
        self.mapp = np.zeros((mapSize, mapSize), dtype='int32',)
        for j in range(1,self.snakeSize):
            self.mapp[1][j] = 1
        
        self.mapp[1][self.snakeSize] = 2

# /**          -->>  Set first direction for each node  <<--  */
        self.builtDirection(mapSize)
        
# /**        -->>  Make first food randomly and set the location for it  <<--  */
        self.x = self.rand()
        self.y = self.rand()
        while (self.checkRand(self.mapp, self.x, self.y)):
            self.x = self.rand()
            self.y = self.rand()
        
        self.mapp[self.x][self.y] = 3
        self.food = self.directions[self.x][self.y]
        
# /**        -->>  Start automatic move the snake  <<--   */
#         self.move()
        self.count = 0
    def nextMove(self, nextDir):
        self.dirr = nextDir
        if (self.dirr != None):
            self.count = 0
            self.head.setDirection(self.dirr)
        else:
            self.count += 1
            return -1



        cur = self.head
        front = cur

        nextt = 0;
        while (True) :
            curDir = cur.getDirection()
            if curDir == self.strRight:
                nextt = cur.getY() + 1
                if (nextt > len(self.mapp) - 1):
#                             next = 0;
                    return -1
                if (self.checkCrash(cur.getX(), nextt, self.mapp)):
                    return -1
                self.swap(self.mapp, cur.getX(), cur.getY(), cur.getX(), nextt)
                self.swapRight(self.directions, cur, front, nextt);
                front = cur
                cur = cur.getPrev()

            elif curDir == self.strLeft:
                nextt = cur.getY() - 1
                if (nextt < 0):
#                              nextt = mapp.length - 1
                    return -1
                if (self.checkCrash(cur.getX(), nextt, self.mapp)):
                    return -1
                self.swap(self.mapp, cur.getX(), nextt, cur.getX(), cur.getY())
                self.swapLeft(self.directions, cur, front, nextt)
                front = cur
                cur = cur.getPrev()

            elif curDir == self.strUP:
                nextt = cur.getX() - 1
                if (nextt < 0):
#                             nextt = mapp.length - 1
                    return -1
                if (self.checkCrash(nextt, cur.getY(), self.mapp)):
                    return -1
                self.swap(self.mapp, cur.getX(), cur.getY(), nextt, cur.getY())
                self.swapUp(self.directions, cur, front, nextt)
                front = cur
                cur = cur.getPrev()

            elif curDir == self.strDown:
                nextt = cur.getX() + 1
                if (nextt > len(self.mapp) - 1):
#                             nextt = 0
                    return -1
                if (self.checkCrash(nextt, cur.getY(), self.mapp)):
                    return -1
                self.swap(self.mapp, nextt, cur.getY(), cur.getX(), cur.getY())
                self.swapRDown(self.directions, cur, front, nextt)
                front = cur
                cur = cur.getPrev()

            else:
                for k in range(len(self.directions)):
                    for j in range(len(self.directions)):
                        if (self.mapp[k][j] == 2):
                            self.head = self.directions[k][j]


# //                        System.out.println(head.getX() + " " + head.getY());
# //                        System.out.println(food.getX() + " " + food.getY());
                if (self.head.getY() == self.food.getY() and self.head.getX() == self.food.getX()) :

                    self.snakeSize += 1

                    self.food.setDirection(self.strNothing)
                    n = self.Node(0, 0)
                    n.setPrev(n)
                    self.food.setPrev(n)
                    front.setPrev(self.food)
                    self.food.setX(front.getX())
                    self.food.setY(front.getY())
                    self.mapp[self.food.getX()][self.food.getY()] = 1

                    cur.setDirection(self.strNothing)

                    self.x = self.rand()
                    self.y = self.rand()
                    while (self.checkRand(self.mapp, self.x, self.y)) :
                        self.x = self.rand()
                        self.y = self.rand()

                    self.mapp[self.x][self.y] = 3
                    self.food = self.directions[self.x][self.y]
                    break 
                else :
                    front.setDirection(self.strNothing);
                    self.directions[front.getX()][front.getY()].setDirection(self.strNothing);
                    cur.setDirection(self.strNothing);
                    break 

        val = self.getNueralInput(self.mapp);

#         for i in range(len(self.mapp)):
#             for j in range(len(self.mapp)):
#                 print(self.mapp[i][j], " ", end='')
#             print("\n")
        return val
#     /**   -->>  This method have thread to move the snake automatically   <<--             */
    def move (self) :
        cout = 0
        dirr = None;
        while (self.snakeSize < len(self.mapp) * len(self.mapp) and cout<30) :
            self.dirr = input()
            if (self.dirr != None):
                cout = 0
                if (self.dirr=="E"):
                    break
                self.head.setDirection(self.dirr)
            else:
                cout += 1
#                 sleep(self.delay)
                continue
            


            cur = self.head
            front = cur

            nextt = 0;
            while (True) :
                curDir = cur.getDirection()
                if curDir == self.strRight:
                    nextt = cur.getY() + 1
                    if (nextt > len(self.mapp) - 1):
#                             next = 0;
                        return
                    if (self.checkCrash(cur.getX(), nextt, self.mapp)):
                        return
                    self.swap(self.mapp, cur.getX(), cur.getY(), cur.getX(), nextt)
                    self.swapRight(self.directions, cur, front, nextt);
                    front = cur
                    cur = cur.getPrev()
                    
                elif curDir == self.strLeft:
                    nextt = cur.getY() - 1
                    if (nextt < 0):
#                              nextt = mapp.length - 1
                        return
                    if (self.checkCrash(cur.getX(), nextt, self.mapp)):
                        return
                    self.swap(self.mapp, cur.getX(), nextt, cur.getX(), cur.getY())
                    self.swapLeft(self.directions, cur, front, nextt)
                    front = cur
                    cur = cur.getPrev()
                    
                elif curDir == self.strUP:
                    nextt = cur.getX() - 1
                    if (nextt < 0):
#                             nextt = mapp.length - 1
                        return
                    if (self.checkCrash(nextt, cur.getY(), self.mapp)):
                        return
                    self.swap(self.mapp, cur.getX(), cur.getY(), nextt, cur.getY())
                    self.swapUp(self.directions, cur, front, nextt)
                    front = cur
                    cur = cur.getPrev()
                    
                elif curDir == self.strDown:
                    nextt = cur.getX() + 1
                    if (nextt > len(self.mapp) - 1):
#                             nextt = 0
                        return
                    if (self.checkCrash(nextt, cur.getY(), self.mapp)):
                        return
                    self.swap(self.mapp, nextt, cur.getY(), cur.getX(), cur.getY())
                    self.swapRDown(self.directions, cur, front, nextt)
                    front = cur
                    cur = cur.getPrev()
                    
                else:
                    for k in range(len(self.directions)):
                        for j in range(len(self.directions)):
                            if (self.mapp[k][j] == 2):
                                self.head = self.directions[k][j]

                    if (self.head.getY() == self.food.getY() and self.head.getX() == self.food.getX()) :

                        self.snakeSize += 1

                        self.food.setDirection(self.strNothing)
                        n = self.Node(0, 0)
                        n.setPrev(n)
                        self.food.setPrev(n)
                        front.setPrev(self.food)
                        self.food.setX(front.getX())
                        self.food.setY(front.getY())
                        self.mapp[self.food.getX()][self.food.getY()] = 1

                        cur.setDirection(self.strNothing)

                        self.x = self.rand()
                        self.y = self.rand()
                        while (self.checkRand(self.mapp, self.x, self.y)) :
                            self.x = self.rand()
                            self.y = self.rand()

                        self.mapp[self.x][self.y] = 3
                        self.food = self.directions[self.x][self.y]
                        break 
                    else :
                        front.setDirection(self.strNothing);
                        self.directions[front.getX()][front.getY()].setDirection(self.strNothing);
                        cur.setDirection(self.strNothing);
                        break 
                        
#             sleep(self.delay)
# //            String mm = "\n\n";
# //            for (int i = 0; i < map.length; i++) {
# //                for (int j = 0; j < map.length; j++) {
# //                    mm = mm + map[i][j] + " ";
# //                }
# //                mm = mm + "\n";
# //            }
# //            System.out.println();
            val = self.getNueralInput(self.mapp);
            
            for i in range(len(self.mapp)):
                for j in range(len(self.mapp)):
                    print(self.mapp[i][j], " ", end='')
                print("\n")
#             writeFile("../resources/java" + name, val);
        
#         System.out.print(" " + head.getDirection() + " " + dir + " ")
#         for (int i = 0; i < map.length; i++) {
#             for (int j = 0; j < map.length; j++) {
#                System.out.print(map[i][j] + " ");
#             }
#             System.out.print("\n");
#         }


    
    def getNueralInput(self, mapp):
        isFrontOpen = 1
        isLeftOpen = 1
        isRightOpen = 1
        isFoodFront = 1
        isFoodLeft = 1
        isFoodRight = 1
        headDir = self.head.getDirection()
        if headDir == self.strRight:
            if (self.head.getY()+1>=len(mapp) or mapp[self.head.getX()][self.head.getY()+1]==1):
                isFrontOpen = 0
            if (self.head.getX()-1<0 or mapp[self.head.getX()-1][self.head.getY()]==1):
                isLeftOpen = 0
            if (self.head.getX()+1>=len(mapp) or mapp[self.head.getX()+1][self.head.getY()]==1):
                isRightOpen = 0
            if (self.head.getY()>=self.food.getY()):
                isFoodFront = 0
            if (self.head.getX()<=self.food.getX()):
                isFoodLeft = 0
            if (self.head.getX()>=self.food.getX()):
                isFoodRight = 0
                
        if headDir == self.strLeft:
            if (self.head.getY()-1<0 or mapp[self.head.getX()][self.head.getY()-1]==1):
                isFrontOpen = 0
            if (self.head.getX()+1>=len(mapp) or mapp[self.head.getX()+1][self.head.getY()]==1):
                isLeftOpen = 0
            if (self.head.getX()-1<0 or mapp[self.head.getX()-1][self.head.getY()]==1):
                isRightOpen = 0
            if (self.head.getY()<=self.food.getY()):
                isFoodFront = 0
            if (self.head.getX()>=self.food.getX()):
                isFoodLeft = 0
            if (self.head.getX()<=self.food.getX()):
                isFoodRight = 0
                    
        if headDir == self.strUP:
            if (self.head.getX()-1<0 or mapp[self.head.getX()-1][self.head.getY()]==1):
                isFrontOpen = 0
            if (self.head.getY()-1<0 or mapp[self.head.getX()][self.head.getY()-1]==1):
                isLeftOpen = 0
            if (self.head.getY()+1>=len(mapp) or mapp[self.head.getX()][self.head.getY()+1]==1):
                isRightOpen = 0
            if (self.head.getX()<=self.food.getX()):
                isFoodFront = 0
            if (self.head.getY()<=self.food.getY()):
                isFoodLeft = 0
            if (self.head.getY()>=self.food.getY()):
                isFoodRight = 0
            
        if headDir == self.strDown:
            if (self.head.getX()+1>=len(mapp) or mapp[self.head.getX()+1][self.head.getY()]==1):
                isFrontOpen = 0
            if (self.head.getY()+1>=len(mapp) or mapp[self.head.getX()][self.head.getY()+1]==1):
                isLeftOpen = 0
            if (self.head.getY()-1<0 or mapp[self.head.getX()][self.head.getY()-1]==1):
                isRightOpen = 0
            if (self.head.getX()>=self.food.getX()):
                isFoodFront = 0
            if (self.head.getY()>=self.food.getY()):
                isFoodLeft = 0
            if (self.head.getY()<=self.food.getY()):
                isFoodRight = 0
                

        return ""+str(isFrontOpen)+"-"+str(isLeftOpen)+"-"+str(isRightOpen) \
                +"-"+str(isFoodFront)+"-"+str(isFoodLeft)+"-"+str(isFoodRight)+"-"+str(self.snakeSize)

#     /**   -->>  These methods work: swap the current Node with the previous;   <<--        */
    def swapUp (self, directions, cur, front, nextt):
        n = self.Node(nextt, cur.getY(), front.getDirection(), cur)
        self.directions[nextt][cur.getY()] = n
    
    def swapRight (self, directions, cur, front, nextt):
        n = self.Node(cur.getX(), nextt, front.getDirection(), cur)
        self.directions[cur.getX()][nextt] = n
    
    def swapLeft (self, directions, cur, front, nextt):
        n = self.Node(cur.getX(), nextt, front.getDirection(), cur)
        self.directions[cur.getX()][nextt] = n
    
    def swapRDown (self, directions, cur, front, nextt):
        n = self.Node(nextt, cur.getY(), front.getDirection(), cur)
        self.directions[nextt][cur.getY()] = n
    

#     /**   -->>  This method will swap the nodes in the map array   <<--                   */
    def swap (self, mapp, i, j, x, y):
        tmp = mapp[i][j]
        mapp[i][j] = mapp[x][y]
        mapp[x][y]=tmp

#     /**   -->>  This method will set first direction for each node  <<--                  */
    def builtDirection (self, size) :
        self.directions = [self.Node() for i in range(size)]
        for i in range(size):
            self.directions[i] = [self.Node() for i in range(size)]
            
        for i in range(size):
            for j in range(size):
                if (self.mapp[i][j]==1):
                    self.directions[i][j] = self.Node(i,j)
                    self.directions[i][j].setDirection(self.strRight)
        
                elif (self.mapp[i][j]==2):
                    self.directions[i][j] = self.Node(i,j)
                    self.directions[i][j].setDirection(self.strRight)
                    self.head = self.directions[i][j]
                
                else :
                    self.directions[i][j] = self.Node(i,j)
                    self.directions[i][j].setDirection(self.strNothing)
                
            
        for i in range(size):
            for j in range(size):
                prev = j-1;
                if (prev<0):
                    prev = len(self.directions)-1;
                if (not self.directions[i][j].getDirection()==self.strNothing):
                    self.directions[i][j].setPrev(self.directions[i][prev]);
            

#     /**   -->>  Class node  <<--    */
    class Node :
        
        def __init__(self, x=0, y=0, direction=None, prev=None):
            
            self.nodeX = x                 #//  The number of the row of this node int the map arr
            self.nodeY = y                 #//  The number of the column of this node int the map arr
            self.direction = direction     #//  Jahat or the direction of each node
            self.prev = prev               #//  Address of the previous node
        
#         //        -->>  Setters  <<--
        def setDirection(self, direction) :
            self.direction = direction
        
        def setPrev(self, prev) :
            self.prev = prev
        
        def setX(self, nodeX) :
            self.nodeX = nodeX
        
        def setY(self, nodeY) :
            self.nodeY = nodeY
        
#         //         -->>  Getters  <<--
        def getPrev(self) :
            return self.prev
        
        def getDirection(self) :
            return self.direction
        
        def getX(self) :
            return self.nodeX
        
        def getY(self) :
            return self.nodeY
        
#     /**   -->>  Build new random number between 0 to map array length  <<--             */
    def rand(self) :
        r = int(np.random.randint(len(self.mapp)))
        if (r < 0):
            r = -r
        return r

#     /**   -->>  Check the random number to not be on the snake  <<--                    */
    def checkRand(self, mapp, x, y):
        return mapp[x][y]!=0

#     /**   -->>  Check the crash the snake whit the snake   <<--                         */
    def checkCrash(self, x1,  y1, mapp):
        return mapp[x1][y1] == 1;
```

```
def Crossover(model1, model2):
  Newmodel1 = newNNModel()
  Newmodel2 = newNNModel()
  we1 = model1.get_weights()
  we2 = model2.get_weights()
  we1N = list()
  we2N = list()
  c = 0
  for w1, w2 in zip(we1, we2):
    start = int(np.random.randint(len(w1)))
    end = len(w1)//2
    if end < start:
      t = start
      start = end
      end = t

#         print(start, ' ', end , '\n\n')

    a = np.ndarray(w1.shape)
    b = np.ndarray(w1.shape)
    for i in range(start):
      a[i] = w1[i]
      b[i] = w2[i]

    for i in range(start, end):
      a[i] = w2[i]
      b[i] = w1[i]
    for i in range(end, len(w1)):
      a[i] = w1[i]
      b[i] = w2[i]

    we1N.append(a)
    we2N.append(b)
    c += 1

  Newmodel1.set_weights(we1N)
  Newmodel2.set_weights(we2N)
  # model1.set_weights(we1N)
  # model2.set_weights(we2N)
  return Newmodel1, Newmodel2
  
def Mutation(model):
  we = model.get_weights()
  
  index = int(np.random.randint(len(we)))
  if( index == 0 or index == 2 or index == 4 ):
    index1 = int(np.random.randint(len(we[index])))
    index2 = int(np.random.randint(len(we[index][0])))
    we[index][index1][index2] += 0.05 
  else :
    index1 = int(np.random.randint(len(we[index])))
    we[index][index1] += 0.05 

  if( index == 0 or index == 2 or index == 4 ):
    index1 = int(np.random.randint(len(we[index])))
    index2 = int(np.random.randint(len(we[index][0])))
    we[index][index1][index2] -= 0.05 
  else :
    index1 = int(np.random.randint(len(we[index])))
    we[index][index1] -= 0.05 
  
  model.set_weights(we)
  return model
```
